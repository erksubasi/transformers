## Import

```python

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from pprint import pprint
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

### Some alternative models

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
model = SentenceTransformer('all-MiniLM-L12-v2')
```

#Use model to generate embeddings and check similarities based on those

```python
sentences = ['I want to take some risks and make as much money as possible in a short time frame',
             'I am rather risk-averse and when it comes to investing I believe in the tortoise beats the hare.',
             'One has to balance the risk and return in investing, I am personally tilted towards controlling the former.']

sentence_embeddings = model.encode(sentences)
```

```python
sentence_embeddings.shape
```

```python
print(f"""Similarity between 1st and 2nd is {cosine_similarity(sentence_embeddings[0].reshape(1, -1),
                sentence_embeddings[1].reshape(1, -1))[0][0]:.3}""")

print(f"""Similarity between 1st and 3rd is {cosine_similarity(sentence_embeddings[0].reshape(1, -1),
                sentence_embeddings[2].reshape(1, -1))[0][0]:.3}""")

print(f"""Similarity between 2nd and 3rd is {cosine_similarity(sentence_embeddings[1].reshape(1, -1),
                sentence_embeddings[2].reshape(1, -1))[0][0]:.3}""")
```
